```python
def marks_category(marks):
    if marks>=80:
        return 'Good'
    elif 70<=marks<80:
        return 'Avg'
    else:
        return 'Poor'
```

```python
# marks_category(65)
```

```python
# df['col'].apply(func) -> built-in,user-defined,lambda
df['marks_category'] = df['marks'].apply(marks_category)
```

```python
def name_initials(name):
    return name[0].capitalize()
```

```python
name_initials('anoop')
```

```python
df['initials'] = df['first_name'].apply(name_initials)
```

```python
def months(date):
    month = pd.to_datetime(date).strftime('%B')
    return month
```

```python
df['months'] = df['dob'].apply(months)
```
